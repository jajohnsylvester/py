## 2. Creating Custom Research Tools

Research agents need specialized tools to gather and process information. Let's create some essential research tools using the `@function_tool` decorator.

```python
@function_tool
def search_academic_papers(query: str, max_results: int = 5) -> str:
    """
    Search for academic papers on a given topic.
    
    Args:
        query: The research query or topic
        max_results: Maximum number of results to return
    
    Returns:
        A formatted string with paper titles and abstracts
    """
    # Simulated academic paper search
    # In a real implementation, you would connect to APIs like arXiv, PubMed, or Google Scholar
    papers = [
        {
            "title": f"Advanced Studies in {query}: A Comprehensive Review",
            "authors": "Smith et al.",
            "year": 2024,
            "abstract": f"This paper provides a comprehensive review of recent advances in {query}..."
        },
        {
            "title": f"Machine Learning Applications for {query}",
            "authors": "Johnson & Lee",
            "year": 2024,
            "abstract": f"We explore how machine learning can be applied to {query} problems..."
        },
        {
            "title": f"Future Directions in {query} Research",
            "authors": "Chen et al.",
            "year": 2023,
            "abstract": f"This paper outlines promising future research directions in {query}..."
        }
    ]
    
    results = []
    for paper in papers[:max_results]:
        results.append(
            f"**{paper['title']}**\n"
            f"Authors: {paper['authors']} ({paper['year']})\n"
            f"Abstract: {paper['abstract']}\n"
        )
    
    return "\n".join(results) if results else "No papers found for the given query."

@function_tool
def analyze_data_trends(topic: str, timeframe: str = "last 5 years") -> str:
    """
    Analyze data trends related to a research topic.
    
    Args:
        topic: The topic to analyze
        timeframe: The timeframe for trend analysis
    
    Returns:
        Analysis of trends and patterns
    """
    analysis = f"""
**Trend Analysis for {topic}**

Timeframe: {timeframe}

Key Findings:
1. **Growing Interest**: Research publications on {topic} have increased by 150% over the {timeframe}
2. **Emerging Subtopics**: New focus areas include sustainable approaches and AI integration
3. **Geographic Distribution**: Major research centers in North America (40%), Europe (35%), and Asia (25%)
4. **Industry Adoption**: Commercial applications have grown significantly in healthcare and finance sectors
5. **Future Outlook**: Experts predict continued growth with emphasis on ethical considerations

Data Sources: Academic databases, industry reports, patent filings
    """
    return analysis

@function_tool
def get_expert_opinions(topic: str, num_experts: int = 3) -> str:
    """
    Get expert opinions on a research topic.
    
    Args:
        topic: The topic to get expert opinions on
        num_experts: Number of expert opinions to include
    
    Returns:
        Compiled expert opinions and insights
    """
    experts = [
        {
            "name": "Dr. Sarah Mitchell",
            "affiliation": "MIT",
            "opinion": f"{topic} represents one of the most promising areas of research today. The potential applications are vast."
        },
        {
            "name": "Prof. James Chen",
            "affiliation": "Stanford University",
            "opinion": f"We're seeing breakthrough developments in {topic}. The next decade will be transformative."
        },
        {
            "name": "Dr. Maria Rodriguez",
            "affiliation": "Oxford University",
            "opinion": f"The interdisciplinary nature of {topic} is what makes it so exciting. Collaboration is key."
        }
    ]
    
    opinions = []
    for expert in experts[:num_experts]:
        opinions.append(
            f"**{expert['name']}** ({expert['affiliation']}):\n"
            f"\"{expert['opinion']}\""
        )
    
    return "\n\n".join(opinions)

print("Research tools created successfully!")
```

## 3. Building the Research Agent

Now let's create our research agent with custom instructions and attach our research tools.

```python
# Create the research agent with specialized instructions
research_agent = Agent(
    name="Research Assistant",
    instructions="""
    You are an expert research assistant specializing in comprehensive analysis and information gathering.
    
    Your approach to research:
    1. Start by understanding the research question thoroughly
    2. Search for relevant academic papers and studies
    3. Analyze trends and patterns in the data
    4. Gather expert opinions when available
    5. Synthesize findings into clear, actionable insights
    
    Always provide:
    - Evidence-based conclusions
    - Multiple perspectives on the topic
    - Clear citations and sources
    - Practical recommendations
    
    Be thorough, objective, and critical in your analysis.
    """,
    model="gpt-4o-mini",  # You can also use "gpt-4o" for more advanced capabilities
    tools=[
        search_academic_papers,
        analyze_data_trends,
        get_expert_opinions
    ],
    model_settings=ModelSettings(
        temperature=0.7,  # Balance between creativity and consistency
        max_tokens=2000   # Allow for comprehensive responses
    )
)

print(f"Research Agent '{research_agent.name}' created with {len(research_agent.tools)} tools")
```

## 5. Basic Tracing with OpenAI Agents SDK

Tracing helps us monitor and debug our agent's execution. Let's see how to use basic tracing features.

```python
# Example: Using trace context manager to group operations
print("=" * 50)
print("Example: Multi-step Research with Tracing")
print("=" * 50)

async def traced_research_workflow():
    """Demonstrate tracing with multiple agent calls"""
    
    with trace("Multi-step Research Workflow") as research_trace:
        print(f"Started trace: {research_trace.id}")
        
        # Step 1: Initial research
        print("\nðŸ“š Step 1: Initial Research...")
        result1 = await Runner.run(
            research_agent, 
            "What are the key challenges in artificial intelligence research?",
            run_config=RunConfig(workflow_name="AI Research - Initial")
        )
        
        # Step 2: Follow-up analysis
        print("\nðŸ” Step 2: Follow-up Analysis...")
        result2 = await Runner.run(
            research_agent,
            f"Based on these challenges: {result1.final_output[:200]}..., what are the most promising solutions?",
            run_config=RunConfig(workflow_name="AI Research - Solutions")
        )
        
        print("\n" + "=" * 60)
        print("ðŸ“‹ RESEARCH WORKFLOW RESULTS")
        print("=" * 60)
        print("\nðŸ” Initial Research Findings:")
        print(result1.final_output[:500] + "...")
        print("\nðŸ’¡ Promising Solutions:")
        print(result2.final_output[:500] + "...")
        
        return result1, result2

# Run the traced workflow
# Uncomment to run async version:
# import asyncio
# results = await traced_research_workflow()

print("Traced research workflow example prepared (uncomment to run async version)")
```

## 6. Synchronous Tracing Example

For simpler use cases, here's how to do basic tracing with synchronous calls.

```python
# Synchronous tracing example
print("=" * 50)
print("Synchronous Research with Tracing")
print("=" * 50)

def sync_research_workflow():
    """Simple synchronous research with tracing configuration"""
    
    # Configure tracing for the research session
    research_config = RunConfig(
        workflow_name="Synchronous Research Session",
        trace_include_sensitive_data=False,
        group_id="research_session_001"  # Group related traces together
    )
    
    topics = [
        "machine learning trends",
        "climate change solutions",
        "space exploration advances"
    ]
    
    results = []
    
    for i, topic in enumerate(topics, 1):
        print(f"\nðŸ” Researching topic {i}/3: {topic}")
        
        # Each research call will be traced under the same workflow
        result = Runner.run_sync(
            research_agent,
            f"Provide a brief overview of recent developments in {topic}",
            run_config=research_config
        )
        
        results.append({
            "topic": topic,
            "summary": result.final_output[:200] + "..."
        })
        
        print(f"âœ… Completed research on {topic}")
    
    return results

# Run the synchronous research
research_results = sync_research_workflow()

print("\n" + "=" * 60)
print("ðŸ“Š RESEARCH SESSION SUMMARY")
print("=" * 60)

for i, result in enumerate(research_results, 1):
    print(f"\n{i}. **{result['topic'].title()}**")
    print(f"   {result['summary']}")

print(f"\nâœ… Completed research on {len(research_results)} topics")
```

## 8. Streaming Research with Tracing

For longer research tasks, we can stream the response while maintaining tracing capabilities.

```python
import asyncio

async def stream_research_with_tracing(agent, query, workflow_name="Streaming Research"):
    """Stream research responses with tracing"""
    print(f"Research Query: {query}")
    print("\nStreaming Response with Tracing:")
    print("-" * 50)
    
    # Configure streaming with tracing
    stream_config = RunConfig(
        workflow_name=workflow_name,
        trace_include_sensitive_data=False
    )
    
    # Run the agent with streaming
    result = Runner.run_streamed(agent, query, run_config=stream_config)
    
    # Stream the events as they come
    full_response = ""
    async for event in result.stream_events():
        if event.type == "raw_response_event":
            # Print each chunk as it arrives
            chunk = event.data.delta
            print(chunk, end="", flush=True)
            full_response += chunk
    
    print("\n" + "-" * 50)
    print(f"\nâœ… Streaming complete! Workflow: {workflow_name}")
    return full_response

# Example streaming research query
streaming_query = """
Conduct research on the future of renewable energy:
1. Current state of technology
2. Major challenges
3. Breakthrough innovations expected
4. Timeline for widespread adoption
"""

print("Streaming research example prepared.")
print("To run: await stream_research_with_tracing(research_agent, streaming_query)")

# Uncomment to run the streaming example:
# response = await stream_research_with_tracing(research_agent, streaming_query)
```

## 9. Best Practices for Research Agents

Here are key best practices when building research agents with the OpenAI Agents SDK.

```python
# Best Practices Examples

# 1. Error Handling in Tools
@function_tool
def safe_research_tool(query: str) -> str:
    """
    A research tool with proper error handling.
    """
    try:
        # Validate input
        if not query or len(query.strip()) < 3:
            return "Error: Query must be at least 3 characters long"
        
        # Your research logic here
        result = f"Research results for: {query}"
        return result
        
    except Exception as e:
        return f"Error during research: {str(e)}"

# 2. Structured Configuration Management
def create_research_config(workflow_name: str, include_sensitive: bool = False) -> RunConfig:
    """Create standardized research configuration"""
    return RunConfig(
        workflow_name=workflow_name,
        trace_include_sensitive_data=include_sensitive,
        trace_metadata={
            "created_at": datetime.now().isoformat(),
            "agent_type": "research_agent",
            "version": "1.0"
        }
    )

# 3. Research Session Manager
class ResearchSession:
    """Manage research sessions with consistent tracing"""
    
    def __init__(self, session_id: str, agent: Agent):
        self.session_id = session_id
        self.agent = agent
        self.results = []
    
    def research(self, query: str, workflow_suffix: str = "") -> str:
        """Conduct research with consistent configuration"""
        workflow_name = f"Research Session {self.session_id}"
        if workflow_suffix:
            workflow_name += f" - {workflow_suffix}"
        
        config = create_research_config(workflow_name)
        result = Runner.run_sync(self.agent, query, run_config=config)
        
        # Store result
        self.results.append({
            "timestamp": datetime.now().isoformat(),
            "query": query,
            "response": result.final_output
        })
        
        return result.final_output
    
    def get_session_summary(self) -> Dict[str, Any]:
        """Get summary of the research session"""
        return {
            "session_id": self.session_id,
            "total_queries": len(self.results),
            "results": self.results
        }

# Example usage of best practices
print("Best Practices Implementation:")
print("=" * 40)

# Create a research session
session = ResearchSession("demo_001", research_agent)

# Conduct research
print("\nðŸ”¬ Conducting research with best practices...")
result = session.research(
    "What are the latest trends in quantum computing applications?", 
    "Quantum Trends"
)

print(f"âœ… Research completed: {result[:100]}...")
print(f"ðŸ“Š Session has {len(session.results)} completed research queries")

print("\nâœ… Best practices examples loaded!")
```
